```python
# Скачиваем код с репозитория
import requests

gh_code = 'https://raw.githubusercontent.com/IvanRychkov/olap_tools/master/partitioner.py'

with open('partitioner.py', 'w+') as p:
  p.write(requests.get(gh_code).text)

from partitioner import execute
```

### Задаём параметры
Впиши сюда свои параметры генерации партиций и нажми Shift + Enter.

Сгенерированные партиции появятся ниже и в файле output.txt во вкладках меню слева. Их можно отправить себе на внутреннюю почту.

```python
# Таблица
table = 'rep70061_fct_transaction'
# База
db = 'prod_v_rep'
# Поле, по которому делается срез
slice_column = 'transaction_dt'
# Дата начала первой партиции
date_begin = '2019-01-01'
# Последняя дата
date_end = '2021-12-31'
# Размер партиции: month или day
partition_size = 'month'

##############################################
result = execute(partition_size, date_begin, date_end, db, table, slice_column)
with open('output.txt', 'w+') as file:
    file.write(result)
print(result)
```
